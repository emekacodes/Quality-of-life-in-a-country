```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.worlddata.info/quality-of-life.php').content
soup = BeautifulSoup(page, 'html.parser')
rows = soup.find_all('tr')
```

```python
dataframe = []
for i in rows[1:]:
    rank = i.find_all('td')[0].contents[0]
    country = i.find('a').contents[0]
    stability = i.find_all('td')[2].contents[0]
    rights = i.find_all('td')[3].contents[0]
    health = i.find_all('td')[4].contents[0]
    safety = i.find_all('td')[5].contents[0]
    climate = i.find_all('td')[6].contents[0]
    costs = i.find_all('td')[7].contents[0]
    popularity = i.find_all('td')[8].contents[0]
    total = i.find_all('td')[9].contents[0]
    dataframe.append({
        'rank' : rank,
        'country' : country,
        'stability' : stability,
        'rights' : rights,
        'health' : health,
        'safety' : safety,
        'climate' : climate ,
        'costs' : costs,
        'popularity' : popularity,
        'total' : total
    })
df = pd.DataFrame(dataframe)
```

```python
df.to_csv('quality_extracted.csv' ,index = False)
```
